```python
columns = list(df_total)
columns
```

```python
fig = px.line(df_total,y=columns)
fig.update_xaxes(tickangle=90,tickvals = np.arange(0,32),ticktext = df_2005_T.index)
fig.show()
```

```python
repo_url = 'https://raw.githubusercontent.com/angelnmara/geojson/master/mexicoHigh.json'
mx_regions_geo = requests.get(repo_url).json()

for i in columns: 
    fig = px.choropleth(
        data_frame=df_total,
        geojson = mx_regions_geo,
        locations = df_2005_T.index,
        featureidkey = 'properties.name',
        color = i,
        color_continuous_scale = 'Reds',
        scope = "north america",
                        )
    fig.update_geos(showcountries = True, showcoastlines =True, showland = True,fitbounds = "locations")

    fig.update_layout(
        title_text = f'Número de Donaciones de Sangre en México ({i})',
        font = dict(
            #family = 'Courier New, monospace",
            family = "Ubuntu",
            size = 18,
            color = "#7f7f7f"  
        ),
        

    )

    fig.show()
    fig.write_image(f"fig{i}.png")
```
