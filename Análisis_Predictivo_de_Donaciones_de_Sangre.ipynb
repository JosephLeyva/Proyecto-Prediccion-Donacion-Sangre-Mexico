# Linear Regression

```python
x = np.arange(2005,2020).reshape(-1,1)
c = 0
for i in df_total_T.columns:
  y = df_total_T.iloc[:, c].values.reshape(-1, 1)
  model = LinearRegression().fit(x, y)
  plt.plot(x,y)
  plt.title(f"Número de Donaciones por Año en {i}")
  plt.show()
  c+=1
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Aqui se pone el año en el que se desea predecir las donaciones de sangre para 'x' estado
y_pred = model.predict([[2020]])
y_pred
```

#Lasso

```python
from sklearn.linear_model import Lasso

# Instantiate a lasso regressor: lasso
lasso = Lasso(alpha=0.4)

y = df_total_T.iloc[:, 0].values.reshape(-1, 1)

# Fit the regressor to the data
lasso.fit(x,y)
```
